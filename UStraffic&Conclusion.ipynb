# US Airline Traffic Analysis
Using a csv file [from Kaggle](https://www.kaggle.com/datasets/yyxian/u-s-airline-traffic-data/data), there is data which we could use from 2019 to 2021 to analyze how the airline traffic potentially performed over the ubiquitous COVID-19 era. The csv file is already present in the project folder and we will use pandas to store the data, and visualize it via plotting to evidently highlight impact of COVID-19 on the number of passengers and the number of flights in the US. 

Side note: GitHub does not accept large file :/ had to make a new file within my folder

```python
import csv
import matplotlib.pyplot as plt

# Open the CSV file and read the content
with open('US_air_traffic.csv', 'r') as file:
    content = csv.reader(file)
    next(content)  # Skip the header

    # Initialize lists to store the data
    year = []
    month = []

    num_domestic_flights = []
    num_international_flights = []

    num_domestic_passengers = []
    num_international_passengers = []

    # Read each row in the CSV file
    for row in content:
        # Filter rows for the years 2019 to 2021
        if int(row[0]) >= 2019 and int(row[0]) <= 2022:
            year.append(f'{row[0]} - {row[1]}')  # Combine year and month for better x-axis labels
            num_domestic_flights.append(int(row[5].replace(",", "")))  # Convert to integer and remove commas
            num_international_flights.append(int(row[6].replace(",", ""))) 

            num_domestic_passengers.append(int(row[2].replace(",", "")))
            num_international_passengers.append(int(row[3].replace(",", "")))
            

# Plotting the data
plt.figure(figsize=(14, 6))
plt.plot(year, num_domestic_flights, label= 'Domestic Flights', marker='o')
plt.plot(year, num_international_flights, label = 'International Flights', marker = 'o')
plt.xlabel('Time')
plt.ylabel('Number of Flights')
plt.title('Number of Domestic & International Flights from 2019 to 2021 from the US')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 6))
plt.plot(year, num_domestic_passengers, label= 'Domestic Passengers', marker='o')
plt.plot(year, num_international_passengers, label = 'International Flights', marker = 'o')
plt.xlabel('Time')
plt.ylabel('Number of Passengers (in tens of millions)')
plt.title('Number of Domestic & International Passengers from 2019 to 2021 from the US')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Insights about US airline transportation from 2019 to 2021
The first graph shows the number of domestic and international flights in the US, and the second graph shows the number of domestic and international passengers in the US, both with respect to time. It can be evidently proven that COVID-19 has made an influential impact on the aviation industry. Over the time period from late 2019 to early to mid 2020, the number of flights and the number of passengers, regardless of international or domestic, had sharply decreased. This trend is particularly due to the implementation of travel restrictions, lockdown measures, and a significant decline in passenger demand due to health and safety concerns. As the pandemic spread, airlines faced unprecedented challenges, leading to reduced flight schedules, grounded fleets, and a substantial drop in passenger volumes. The decrease in air travel was a direct consequence of the global efforts to contain the virus and the public's cautious approach to avoid non-essential travel during the peak periods of the pandemic.

Fast forward to 2022 and onwards, the numbers are in a recovery because COVID-19 is no longer a current pandemic due to the vaccines biomedically engineered by companies like Pfizer or Moderna. This has boosted public confidence because they would potentially be never or less be affected by COVID-19 if it exists in any particular amount. However, it is still worth noting how a pandemic would minimize travelling due to safety concerns and how that would affect companies by itself, as discussed earlier. 

# Conclusion

The COVID-19 pandemic had a profound impact on the aerospace industry, affecting both suppliers of aerospace materials and airline firms. This disruption was particularly evident in the equity value of stocks and overall revenue.

- **Decreased Demand and Operational Challenges:**
The pandemic led to a sharp decline in aerospace material demand as airlines reduced operations drastically. This downturn severely impacted revenue streams for suppliers.

- **Record-Low Passenger Numbers and Financial Strain:**
Global lockdowns and health precautions in 2020 caused unprecedented drops in passenger numbers, straining airlines financially. Many airlines reported negative net incomes, affecting their financial health and investor confidence.

- **Investor Response and Market Dynamics:**
Investors reacted to airlines' financial instability by selling off stocks, intensifying industry-wide equity declines. Uncertain recovery timelines and ongoing travel restrictions weighed heavily on investor sentiment.

- **Supply Chain Disruptions and Recovery Efforts:**
The aerospace industry also faced supply chain disruptions. Suppliers adjusted production and inventory management amid fluctuating demand. Recovery hinged on vaccine distribution, government policies, and restoring consumer confidence in travel safety.
