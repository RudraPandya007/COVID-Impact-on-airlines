### Supply of Aerospace Materials

The aerospace industry, simply put Airline industry, is a highly complex and demands materials and components that meet stringent standards for performance, safety, and durability. The supply of aerospace materials involves a range of products such as aircraft engines, avionics, structural components, and various specialized materials. To begin with, let's focus the data analysis on four supply companies which are Rolls-Royce Holdings plc, Airbus SE, General Electric, and Boeing Company.

### Data about prices during COVID era
To comprehend the impact of COVID-19 on the aviation industry, it is essential to analyze the stock prices of key companies from 2019 to 2021. Displaying the stock price graphs of major aerospace companies such as Rolls-Royce Holdings plc, Airbus SE, General Electric, and Boeing Co during this period will illustrate the significant financial damage inflicted by the pandemic. These graphs will reveal the stark declines in stock values as global travel restrictions and lockdowns severely affected airline operations and demand for new aircraft. By examining the fluctuations and eventual recoveries, we can gain a clear understanding of the extent of the economic disruption caused by COVID-19 and its impact on the equities of these pivotal companies in the aviation sector. Here are some terms with their respective definition for general familiarity:

- **Opening Price:** Price of the stock when the market opens at 9:30 AM (EST, NYC time)

- **Highest Price:** Highest price the stock has ever reached at that particular date

- **Lowest Price:** Lowest price the stock has ever reached at that particular date

- **Closing Price:** Closing price of the stock when the market closes at 4:00 PM (EST, NYC time)

- **Volume:** The amount of shares bought or sold at that particular date

- **Dividends:** A portion of money paid to shareholders from company's earnings

- **Stock Split:** For example, a 10 for 1 stock split is where existing shareholders' shares are multiplied by 10 and the price of the stock is reduced to divided by 10.

````python
# List of tickers of suppliers of aerospace goods
tickers = ['RYCEY', 'EADSY', 'GE', 'BA']

# Date range
start_date = "2019-12-30"
end_date = "2021-01-02"

# Fetch the data
for ticker in tickers:
    ticker_data = yf.Ticker(ticker).history(start = start_date, end = end_date)

    #Plot it individually
    plt.figure(figsize = (20, 7))
    plt.plot(ticker_data.index, ticker_data['Close'], label = 'Closing Price', color=f'#{random.randint(0, 0xFFFFFF):06x}')

    # Add labels and title
    plt.title(f'Closing Price for {ticker}')
    plt.xlabel('Date')
    plt.ylabel('Price ($)')
    plt.legend()

    # Show plot
    plt.grid(True)
    plt.tight_layout()
    plt.show
````

#### Impact of COVID-19 on Airline Supply Equities

Over the time period from December of 2019 to January of 2021, the major suppliers of aviation materials as Rolls-Royce Holdings plc, Airbus SE, General Electric, and Boeing Co suffered loss of equity value due to the COVID era which was omnipresent in 2020. Many other suppliers as well had a decrease in value and they have a rigorous time because they are hardly able to meet the price standards of the equities of pre-COVID times. 

Over the COVID era, domestic and international travelling both in and out were minimized for the safety of population mitigizing the risk of COVID. Such decrease in travelling would potentially mean that most of the supply is left unused or unrecognized, pessimizing investor confidence, which strongly could have led to decrease in equity prices of various airline suppliers. ### Demand of Aerospace Materials

As discussed, the supply of aerospace materials is backed by companies which specialize in manufacturing of complex designs and engines for the aviation industry. This supply is used by consumers which represent the demand of the materials. For example, Boeing Co would make engines and British Airways would buy it for their airplanes. Likewise, there are dozens of examples of how aviation consumers purchase products from the industry supply to meet their needs. In here, we will focus on three companies which are American Airlines Group Inc, Delta Airlines, and United Airlines.

````python
# List of tickers of suppliers of aerospace goods
tickers = ['AAL', 'DAL', 'UAL']

# Date range
start_date = "2019-12-30"
end_date = "2021-01-02"

# Fetch the data
for ticker in tickers:
    ticker_data = yf.Ticker(ticker).history(start = start_date, end = end_date)

    #Plot it individually
    plt.figure(figsize = (20, 7))
    plt.plot(ticker_data.index, ticker_data['Close'], label = 'Closing Price', color=f'#{random.randint(0, 0xFFFFFF):06x}')

    # Add labels and title
    plt.title(f'Closing Price for {ticker}')
    plt.xlabel('Date')
    plt.ylabel('Price ($)')
    plt.legend()

    # Show plot
    plt.grid(True)
    plt.tight_layout()
    plt.show
````

### Impact of COVID-19 on Airline Consumer Equties
