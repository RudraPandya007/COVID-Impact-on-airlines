### American Airlines earning analysis



In here, we will focus on earnings analysis of American Airlines as a representative of different consumers of aviation supply. We will focus on plotting revenues and costs and identify trends and plausible reasons behind those trends. 

```python
import yfinance as yf
import matplotlib.pyplot as plt

display(yf.Ticker('AAL').financials)
```

```python
revenue_data = {
    '2019-12-31': 45800000000,
    '2020-12-31': 17337000000,
    '2021-12-31': 29882000000,
    '2022-12-31': 48971000000
}

costs_data = {
    '2019-12-31': 27800000000,
    '2020-12-31': 24563000000,
    '2021-12-31': 29855000000,
    '2022-12-31': 39934000000
}

revenue_dates = list(revenue_data.keys())
revenues = list(revenue_data.values())

costs_dates = list(costs_data.keys())
costs = list(costs_data.values())

# Plotting
plt.figure(figsize=(20, 6))
plt.plot(revenue_dates, revenues, marker='o', linestyle='-', color='b', label='Revenue')
plt.plot(costs_dates, costs, marker = 'o', linestyle = '-', color = 'r', label = 'Costs')
plt.plot()

# Add labels and title
plt.title('Revenue & Cost over time')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Revenue ($)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```

### Insights about American Airlines 

During 2019 to 2020, revenues decreased for American Airlines. This is reasonably due to minimal travelling to minimize the risk of COVID-19 for Americans. Due to this action, the company would get less revenue as highlighted in the graph; In addition, the cost of revenue was high in 2020 which led to a net loss for the company in 2020. This can be due to many factors such as maiteaneance costs, engine repairs, fixed operating costs, fuel costs of operating airlines, and many more. 

The point at 2020-12-31 when costs were more than revenue is highlighted in Yfinance's numeric figure of £-7226000000 pounds. In addition, in 2021 at same exact date, the profit was at its minimum, at a nearly breakeven point, standing at £27000000 pounds. The company started to recover after 2021 as COVID-19 risk started to decrease due to vaccine rollouts boosting customer's confidence to travel safely thus increasing revenues.
